```python
import os
import re
```

```python
badcombinations = r'ζπ|ζξ|ζγ|γζ|ζγ|ιιι|οοο|οόο|αεη|ηελ|βξ|ζθ|γπλ|αηη|αηα|ζπγ|πγθ|θαη|ηηο|μδσ|Ǽ|ȃ|ǿ|Ȁ|ȅ|ǹ|Ȇ|ǻ|Ȉ|ȉ|ȇ'
```

```python
os.makedirs('../test_output', exist_ok=True)
for i,file in enumerate(os.listdir('../pergamos')) :
    if not file.endswith('.md') :
        continue
    try:
        with open('../pergamos'+'/'+file, 'r', encoding='utf-8') as infile:
            text = infile.read()
    except Exception as e:
        print(f"Error reading {file}: {e}")
        continue

    error_threshold = 30
    if len(re.findall(badcombinations,text)) < error_threshold :
        t_file = 'good'+file[:-2] + 'txt'
    else :
        t_file = 'bad'+file[:-2] + 'txt'

    output_file_path = os.path.join('../test_output', t_file)
    try:
        with open(output_file_path, 'w', encoding='utf-8') as output_file:
            output_file.write(text)
    except Exception as e:
            print(f"Error writing to {output_file_path}: {e}")
            continue
```
