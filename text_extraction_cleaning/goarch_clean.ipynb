```python
import re
import os
```

```python
os.makedirs('../test_output', exist_ok=True)
for i,file in enumerate(os.listdir('../scraping/goarh_texts')) :
    if not (file.endswith('.html') or file.endswith('.epub')) :
            continue
    try:
        with open('../scraping/goarh_texts'+'/'+file, 'r', encoding='utf-8') as infile:
            text = infile.read()
    except Exception as e:
        print(f"Error reading {file}: {e}")
        continue

    text = re.sub('<[^<]+?>', '', text)
    text = re.sub(r'-|--|•|\+','',text)

    t_file = file[:-4] + 'txt'

    output_file_path = os.path.join('../test_output', t_file)
    try:
        with open(output_file_path, 'w', encoding='utf-8') as output_file:
            output_file.write(text)
    except Exception as e:
            print(f"Error writing to {output_file_path}: {e}")
            continue
```

```python
import pandas as pd

texts = []
with open('metadata','r',encoding='utf-8') as input :
    metadata = input.readlines()
for i,file in enumerate(os.listdir('../test_output')) :
    with open('../test_output'+'/'+file,'r',encoding='utf-8') as input :
        texts.append(input.read())

df = pd.DataFrame({'texts':texts,'katigoria':metadata[:-1]})
df.to_csv('litourgical_texts.csv')
df.to_parquet('litourgical_texts.parquet')
```
