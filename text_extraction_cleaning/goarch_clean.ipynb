```python
import urllib.request
import re
import time
import os
```

```python
os.makedirs('../test_output', exist_ok=True)
for i,file in enumerate(os.listdir('../scraping/goarh_texts')) :
    if not (file.endswith('.html') or file.endswith('.epub')) :
            continue
    try:
        with open('../scraping/goarh_texts'+'/'+file, 'r', encoding='utf-8') as infile:
            text = infile.read()
    except Exception as e:
        print(f"Error reading {file}: {e}")
        continue

    text = re.sub('<[^<]+?>', '', text)
    text = re.sub(r'-|--|•|\+','',text)

    t_file = file[:-4] + 'txt'

    output_file_path = os.path.join('../test_output', t_file)
    try:
        with open(output_file_path, 'w', encoding='utf-8') as output_file:
            output_file.write(text)
    except Exception as e:
            print(f"Error writing to {output_file_path}: {e}")
            continue
```

```python
import pandas as pd

texts = []
links = []
metadata = pd.read_csv('metadata_leitoyrgika.csv')
metadata.drop_duplicates(subset=['link'],inplace=True)
metadata.reset_index(drop=True,inplace=True)
for i,file in enumerate(metadata.iterrows()) :
    user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
    headers={'User-Agent':user_agent,}
    url = file[1][3]
    request=urllib.request.Request(url,None,headers) #The assembled request
    page = response = urllib.request.urlopen(request)
    html_bytes = page.read()
    text = html_bytes.decode("utf-8")
    text = re.sub('<[^<]+?>', '', text)
    text = re.sub(r'-|--|•|\+','',text)
    texts.append(text)
    time.sleep(2.5)
    
df = pd.DataFrame({'texts':texts,'katigoria':metadata['Kathgoria'][:-1],'ypokatigoria':metadata['Ypokathgoria'][:-1],'titlos':metadata['title']})
df.to_csv('litourgical_texts.csv')
df.to_parquet('litourgical_texts.parquet')
```
