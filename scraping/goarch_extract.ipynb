```python
from urllib.request import urlopen
import urllib.request
import re
import time
import os
```

```python
user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
headers={'User-Agent':user_agent,}
url = "https://glt.goarch.org/#02"
request=urllib.request.Request(url,None,headers) #The assembled request
page = response = urllib.request.urlopen(request)
html_bytes = page.read()
html = html_bytes.decode("utf-8")
```

```python
lines = html.splitlines()
links = []
for line in lines :
    links.append(re.findall(r'\".+?\.html\"|\".+?\.epub\"',line))
print(links)
```

```python
for i,link in enumerate(links) :
    flag_html = False
    file_name = "text_"+str(i)
    if link == [] :
        continue
    user_agent = 'Mozilla/5.0 (Windows; U; Windows NT 5.1; en-US; rv:1.9.0.7) Gecko/2009021910 Firefox/3.0.7'
    headers={'User-Agent':user_agent,}
    url = "https://glt.goarch.org/"+link[0][1:-1]
    try :
        request=urllib.request.Request(url,None,headers) #The assembled request
        page = response = urllib.request.urlopen(request)
        html_bytes = page.read()
        if link[0].endswith('.html\"') :
            html = html_bytes.decode("utf-8")
            file_name = file_name+".html"
        else :
            file_name = file_name+".epub"
        with open(file_name, 'w', encoding='utf-8') as output_file:
            output_file.write(html)
    except Exception as e:
        print({e})
    time.sleep(2.5)
```

```python
lines = html.splitlines()
links = []
category = ''
for line in lines :
    if not(re.findall('<FONT COLOR=.+?</FONT>|<font color=.+?</font>',line) == []) :
        category = re.findall('<FONT COLOR=.+?</FONT>|<font color=.+?</font>',line)[0]
        category = re.sub('<[^<]+?>', '', category)
    cat_link = []
    cat_link.append(re.findall(r'\".+?\.html\"|\".+?\.epub\"',line))
    cat_link.append(category)
    links.append(cat_link)
print(links)
```
